```python
from diffusers import StableDiffusionPipeline
import torch

# 모델 이름을 변수로 지정
model_id = "dreamlike-art/dreamlike-photoreal-2.0"
model_id2 = "stabilityai/stable-diffusion-2-1"

# 학습된 모델을 가져오면서 토치 타입도 지정
pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype = torch.float16
)

pipe2 = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype = torch.float16
)

# GPU를 사용하겠다
pipe = pipe.to("cuda")
pipe2 = pipe2.to("cuda")

# 다 됐다고 써논거
print("Model loaded!")
```

```python
img = pipe(
    "sea turtle under the sea"
).images[0]

img
```

```python
# 모델2 이게 훨씬 별로인듯
img2 = pipe2(
    "white sea turtle under the sea"
).images[0]

img2
```
