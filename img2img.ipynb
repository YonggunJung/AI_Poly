```python
import torch
from diffusers import AutoPipelineForImage2Image
from diffusers.utils import load_image, make_image_grid

pipeline = AutoPipelineForImage2Image.from_pretrained(
    "kandinsky-community/kandinsky-2-2-decoder", torch_dtype = torch.float16, use_safetensors = True
)
pipeline.enable_model_cpu_offload()
```

```python
init_image = load_image('/content/Egle.jpg')
init_image
```

```python
prompt = "whale wizard, gandalf, lord of the rings, detailed, fantasy, cute, adorable, Pixar, Disney, 8K"
image = pipeline(prompt, image = init_image).images[0]
make_image_grid([init_image, image], rows = 1, cols = 2)
```

```python
num_cols = 2
num_rows = 2

prompt = ["Egle wizard, gandalf, lord of the rings, detailed, fantasy, cute, adorable, Pixar, Disney, 8K"]
image1 = pipeline(prompt, image1 = init_image).images[0]
image2 = pipeline(prompt, image2 = image1).images[0]
image3 = pipeline(prompt, image3 = image2).images[0]
make_image_grid([init_image, image1, image2, image3], rows = num_rows, cols = num_cols)



```
